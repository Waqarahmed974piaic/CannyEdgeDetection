```python
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\Waqar Ahmed\Desktop\Open_CV\opencv-master\samples\data\lena.jpg",0)
canny=cv2.Canny(img,100,200)

images=[img,canny]
titles=['images','canny']

for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```
